```python
# Tugas
print("="*45)
print("SELAMAT DATANG DI ANGGKRINGAN KEGELAPAN")
print("="*45)
print("Menu Makanan & Minuman:")
print("1. Nasi Kucing         - Rp 6.000")
print("2. Sate Usus           - Rp 8.000")
print("3. Kopi Jahe           - Rp 10.000")
print("4. Teh Panas           - Rp 7.000")
print("5. Pisang Goreng       - Rp 12.000")
print("6. Roti Bakar Cokelat  - Rp 15.000")
print("7. Mie Goreng Spesial  - Rp 18.000")
print("8. Wedang Uwuh         - Rp 9.000")
print("="*45)

# Data menu
menu = {
    1: {"nama": "Nasi Kucing", "harga": 6000},
    2: {"nama": "Sate Usus", "harga": 8000},
    3: {"nama": "Kopi Jahe", "harga": 10000},
    4: {"nama": "Teh Panas", "harga": 7000},
    5: {"nama": "Pisang Goreng", "harga": 12000},
    6: {"nama": "Roti Bakar Cokelat", "harga": 15000},
    7: {"nama": "Mie Goreng Spesial", "harga": 18000},
    8: {"nama": "Wedang Uwuh", "harga": 9000}
}

total_harga = 0
pesanan = []

# Meminta berapa jenis menu yang ingin dibeli
jumlah_menu = int(input("Berapa jenis menu yang ingin Anda pesan? "))

for i in range(jumlah_menu):
    print(f"\nPesanan ke-{i+1}")
    pilihan = int(input("Masukkan nomor menu (1-8): "))

    if pilihan in menu:
        jumlah = int(input(f"Masukkan jumlah {menu[pilihan]['nama']}: "))
        subtotal = menu[pilihan]['harga'] * jumlah
        total_harga += subtotal
        pesanan.append({
            "nama": menu[pilihan]['nama'],
            "jumlah": jumlah,
            "harga": menu[pilihan]['harga'],
            "subtotal": subtotal
        })
    else:
        print("Nomor menu tidak valid, dilewati.")

# Menampilkan ringkasan pesanan
print("\n=== RINGKASAN PESANAN ANDA ===")
for item in pesanan:
    print(f"{item['nama']} x{item['jumlah']} = Rp {item['subtotal']}")
print(f"\nTotal Harga Sebelum Diskon: Rp {total_harga}")

# Menentukan diskon berdasarkan total belanja
if total_harga >= 70000:
    diskon = 0.70
    kategori = "70%"
elif total_harga >= 40000:
    diskon = 0.50
    kategori = "50%"
else:
    diskon = 0.30
    kategori = "30%"

potongan = total_harga * diskon
total_setelah_diskon = total_harga - potongan

print(f"\nDiskon yang Anda dapatkan: {kategori}")
print(f"Potongan harga: Rp {potongan:.0f}")
print(f"Total setelah diskon: Rp {total_setelah_diskon:.0f}")

# Pembayaran
uang_dibayar = int(input("\nMasukkan jumlah uang Anda: Rp "))
if uang_dibayar >= total_setelah_diskon:
    kembalian = uang_dibayar - total_setelah_diskon
    print("\n--- STRUK PEMBAYARAN ---")
    print("ANGKRINGAN KEGELAPAN")
    print("-"*30)
    for item in pesanan:
        print(f"{item['nama']} x{item['jumlah']} = Rp {item['subtotal']}")
    print("-"*30)
    print(f"Total Asli         : Rp {total_harga}")
    print(f"Diskon ({kategori}) : -Rp {potongan:.0f}")
    print(f"Total Bayar        : Rp {total_setelah_diskon:.0f}")
    print(f"Uang Pembayaran    : Rp {uang_dibayar}")
    print(f"Kembalian          : Rp {kembalian}")
    print("-"*30)
    print("Terima kasih telah berkunjung ke Angkringan Kegelapan!")
else:
    kekurangan = total_setelah_diskon - uang_dibayar
    print(f"Maaf, uang Anda kurang sebesar Rp {kekurangan:.0f}.")

print("="*45)
```
