```python
# Soal 7

kapasitas_maksimum = float(input("Masukkan kapasitas maksimum: "))
beban_terpakai = float(input("Masukkan beban yang terpakai: "))

persentase = (beban_terpakai / kapasitas_maksimum) * 100

if persentase <= 60:
  status = "Beban Ringan"
elif persentase <= 85:
  status = "Beban Sedang"
else:
  status = "Beban Berat"

print(f"Persentase beban: {persentase:.2f}%")
print("Status:",status)
```

```python
# Challange

# Fungsi untuk menghitung energi (kWh)
def hitung_energi(daya, jam):
    return (daya * jam) / 1000

# Fungsi untuk menghitung biaya listrik
def hitung_biaya(energi, tarif):
    return energi * tarif

# Fungsi menentukan kategori penggunaan
def kategori(energi):
    if energi < 5:
        return "Hemat "
    elif energi < 15:
        return "Sedang "
    else:
        return "Borors "

# ====== PROGRAM UTAMA ======
print("=== Program Perhitungan Biaya Listrik ===")

ulang = True
while ulang:
    tarif = float(input("\nMasukkan tarif listrik per kWh (contoh: 1500): "))
    jumlah = int(input("Masukkan jumlah perangkat listrik: "))

    total_energi = 0
    total_biaya = 0

    # Perulangan for untuk input data perangkat
    for i in range(jumlah):
        print(f"\nPerangkat ke-{i+1}:")
        nama = input("Nama perangkat: ")
        daya = float(input("Daya (Watt): "))
        jam = float(input("Jam pemakaian per hari: "))

        energi = hitung_energi(daya, jam)
        biaya = hitung_biaya(energi, tarif)

        print(f"{nama} → {energi:.2f} kWh | Rp {biaya:,.2f}")

        total_energi += energi
        total_biaya += biaya

    print("\n=== HASIL TOTAL ===")
    print(f"Total Energi : {total_energi:.2f} kWh")
    print(f"Total Biaya  : Rp {total_biaya:,.2f}")
    print(f"Kategori     : {kategori(total_energi)}")

    # Tanya ingin ulang?
    lagi = input("\nHitung lagi? (y/n): ").lower()
    if lagi != "y":
        ulang = False
print("\nTerima kasih telah menggunakan program ini ")
```
