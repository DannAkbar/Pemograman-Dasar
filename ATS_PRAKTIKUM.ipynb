```python
# SOAL 1
suhu = float(input("Masukkan suhu air pendingin (°C): "))

if suhu <= 90:
  print("STATUS NORMAL")
elif suhu < 100:
  print("STATUS: Warning")
else:
  print("ALERT: SHUTDOWN")
```

```python
# SOAL 3

vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

for i in range(len(vibrasi_mesin)):
  if vibrasi_mesin[i] > 2.0:
    print(f"Warning: Unit {i+1} vibrasi tinggi")
```

```python
# Soal 5

suhu_saat_ini = 60

while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini, "°C")
    suhu_saat_ini += 5
```

```python
# Soal 7

kapasitas_maksimum = float(input("Masukkan kapasitas maksimum: "))
beban_terpakai = float(input("Masukkan beban yang terpakai: "))

persentase = (beban_terpakai / kapasitas_maksimum) * 100

if persentase <= 60:
  status = "Beban Ringan"
elif persentase <= 85:
  status = "Beban Sedang"
else:
  status = "Beban Berat"

print(f"Persentase beban: {persentase:.2f}%")
print("Status:",status)
```

```python
# soal 9

suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]
for suhu in suhu_boiler:
  if suhu <= 100:
    print("Suhu saat ini :", suhu, "°C  (Aman)")
  else:
    print("Peringatan suhu saat ini :", suhu, "°C  (Melebihi Batas!)")
```
